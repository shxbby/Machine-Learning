# Gaussian Blur, Box Blur, Median Blur

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread("snake.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
# Box blur
blurred_img = cv2.blur(img,(10,10))

# Gaussian Blur
blurred_img2 = cv2.GaussianBlur(img,(5,5),0)

# Median Blur
blurred_img3 = cv2.medianBlur(img,3)

# subplot (Row,Col,Sequence)
plt.subplot(2,2,1),plt.imshow(img),plt.title("Original Image"),plt.axis('off')
plt.subplot(2,2,2),plt.imshow(blurred_img),plt.title("Box Image"),plt.axis('off')
plt.subplot(2,2,3),plt.imshow(blurred_img2),plt.title("Gaussian Image"),plt.axis('off')
plt.subplot(2,2,4),plt.imshow(blurred_img3),plt.title("Median Image"),plt.axis('off')
```

# Image Sharpening

```python
#Sharpening Images

import cv2
import numpy as np
import matplotlib.pyplot as plt

img= cv2.imread("snake.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

plt.figure(figsize=(15, 15))

k1 = np.array ([[-1,-1,-1] , [-1, 9, -1] , [-1,-1,-1]])
k2 = np.array ([[1,1,1] , [1, -7, 1] , [1,1,1]])
new_img1 = cv2.filter2D(img , -1 , k1)
new_img2 = cv2.filter2D(img , -1 , k2)
plt.subplot(1,3,1),plt.imshow(img)
plt.axis('off')
plt.subplot(1,3,2),plt.imshow(new_img1)
plt.axis('off')
plt.subplot(1,3,3),plt.imshow(new_img2)
plt.axis('off')
```

# Embossing of Image

```python
#Embossing Images


k1 = np.array ([[0,-1,-1] , [1, 0, -1] , [1,1,0]])
k2 = np.array ([[-1,-1,0] , [-1, 0, 1] , [0,1,1]])

plt.figure(figsize=(15, 15))
new_img1 = cv2.filter2D(img , -1 , k1) +128
new_img2 = cv2.filter2D(img , -1 , k2) + 128

plt.subplot(1,3,1),plt.imshow(img)
plt.axis('off')
plt.subplot(1,3,2),plt.imshow(new_img1)
plt.axis('off')
plt.subplot(1,3,3),plt.imshow(new_img2)
plt.axis('off')
```

# Edge Detection (Laplacian, Canny, Sobel)

```python
#Edge Detection

img= cv2.imread("snake.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
# Sobel
sobelx= cv2.Sobel(img,cv2.CV_64F , 1,0,5)
sobely= cv2.Sobel(img,cv2.CV_64F , 0,1,5)

#lapiacian(img,ddepath)
laplacian_img = cv2.Laplacian(img, cv2.CV_32F)

#canny
canny_img = cv2.Canny(img, 50 , 150 )

plt.figure(figsize=(15, 15))

plt.subplot(1,5,1),plt.imshow(img , cmap="gray")
plt.axis('off')
plt.subplot(1,5,2),plt.imshow(sobelx, cmap="gray")
plt.axis('off')
plt.subplot(1,5,3),plt.imshow(sobely, cmap="gray")
plt.axis('off')

plt.subplot(1,5,4),plt.imshow(laplacian_img, cmap="gray")
plt.axis('off')

plt.subplot(1,5,5),plt.imshow(canny_img, cmap="gray")
plt.axis('off')
```

# Erode and Dilate Images

```python
#Erode and Dilate Images

img = cv2.imread("in.jpg")
k1 = np.ones((5,5))
erode_image = cv2.erode(img,k1,iterations = 2)
dilate_image = cv2.dilate(img,k1,iterations = 2)

plt.subplot(1,3,1), plt.imshow(img)
plt.subplot(1,3,2), plt.imshow(erode_image)
plt.subplot(1,3,3), plt.imshow(dilate_image)
```

# Erode&Dialte image

```python

img=cv2.imread("snake.jpg")
k1=np.ones((5,5))
erode_img=cv2.erode(img,k1,iterations=1)
dialte_img=cv2.dilate(img,k1,iterations=2)

plt.subplot(1,3,1),plt.imshow(img),plt.title("orginal"),plt.axis('off')
plt.subplot(1,3,2),plt.imshow(erode_img),plt.title("erode_img"),plt.axis('off')
plt.subplot(1,3,3),plt.imshow(dialte_img),plt.title("dialte_img")
plt.axis('off')
```

```python
img=cv2.imread("tree.jpg")
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img_YUV=cv2.cvtColor(img,cv2.COLOR_RGB2YUV)
img_YUV[:,:,0]=cv2.equalizeHist(img_YUV[:,:,0])
final_img=cv2.cvtColor(img_YUV,cv2.COLOR_YUV2BGR)

plt.subplot(1,2,1),plt.imshow(img),plt.axis('off')
plt.subplot(1,2,2),plt.imshow(final_img),plt.axis('off')
```

```python
#fleading an image
img=cv2.imread("tree.jpg")
#convert the image in RGB format
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
#convert the image into YUV format
img_YUV=cv2.cvtColor(img,cv2.COLOR_BGR2YUV)
#split the image
img_YUV[:,:,0]=cv2.equalizeHist(img_YUV[:,:,0])
#Convert YUV image into Color
final=cv2.cvtColor(img_YUV,cv2.COLOR_YUV2BGR)
plt.imshow(final)
plt.axis('off')
```

# Contrast on Grey scale

```python
#Contrast on Grey scale


img= cv2.imread("tree.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img_YUV = cv2.cvtColor(img,cv2.COLOR_BGR2YUV)
img_YUV[:,:,0] = cv2.equalizeHist(img_YUV[:,:,0])
final_img = cv2.cvtColor(img_YUV,cv2.COLOR_YUV2BGR)

plt.subplot(1,2,1),plt.imshow(img)
plt.subplot(1,2,2),plt.imshow(final_img)
```
